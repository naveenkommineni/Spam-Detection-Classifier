```python
a=vectorizer=CountVectorizer(max_df=.85,stop_words=stop_words,max_features=1000,token_pattern='[a-zA-Z][a-zA-Z]{2,}')
a=vectorizer.fit_transform(sdata['Text'])
a
a.toarray()
```

```python
feature_names=vectorizer.get_feature_names()
feature_names
```

```python
features=pd.DataFrame(a.toarray(),columns=vectorizer.get_feature_names())
features
```

```python
X=features
```

```python
Y=sdata['map']
```

```python
from sklearn import svm
```

```python
svm_model=svm.SVC()
```

```python
svm.SVC()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=.4,random_state=1)
```

```python
svm_model.fit(X_train,Y_train)
```

```python
Ypred=svm_model.predict(X_test)
```

```python
from sklearn import metrics
metrics.accuracy_score(Y_test,Ypred)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
param_grid= { "kernel"      : ['linear','rbf'],    
               "C"   : [4,11,15,13,21,32 ],  
             "gamma":[0.1,1,1.4,5.2,10],
             "class_weight":[{0:2,1:1},{0:4,1:3},{0:6,1:4}]
            }
```

```python
stree=RandomizedSearchCV(svm_model,param_grid,scoring='precision',verbose=2,cv=3)
```

```python
stree.fit(X_train,Y_train)
```

```python
pred=stree.predict(X_test)
```

```python
metrics.accuracy_score(Y_test,pred)
```

```python
stree.best_params_
```

```python
stree=RandomizedSearchCV(svm_model,param_grid,scoring='recall',verbose=2,cv=3)#sensitivity
```

```python
stree.fit(X_train,Y_train)
```

```python
Pred=stree.predict(X_test)
```

```python
metrics.recall_score(Y_test,Pred)
```

```python
metrics.precision_score(Y_test,Pred)
```
